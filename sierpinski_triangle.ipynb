```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
```

```python
def next_point(current_point):
    tri_corner = [
        np.array((0, np.sqrt(3))),
        np.array((1, 0)),
        np.array((-1, 0))
    ][np.random.choice([0, 1, 2])]
    
    return (tri_corner + current_point)/2

def tri_points():
    pt = np.array((0, 1))
    while True:
        yield pt
        pt = next_point(pt)
        
def draw(num_iterations):
    gen = tri_points()
    fig, ax = plt.subplots(figsize=(5, 5))
    for i in tqdm(range(num_iterations)):
        ax.plot(*next(gen), marker='o', markersize=2)

    plt.axis('off')
    plt.show()
    fig.clear()
    plt.close()
```

```python
draw(100)
```

```python
draw(1000)
```

```python
draw(10000)
```
